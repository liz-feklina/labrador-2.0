```python
def write_news(link, filename):
    r = requests.get(link, cookies=cookies).text
    r = r.replace('\n', '')
    pattern = '<div class="entry">(.+?)</div>'
    news = re.findall(pattern, r)[0]
    #replace('&#8212;', '—')
    news = news.replace('<p>', '\n')
    news = news.replace('<br />', '\n')
    news = re.sub('<.+?>', '', news)
    with open(filename, "a", encoding="utf-8") as f:
        f.write(news)

#text = write_news('http://sp-18.ru/2018/02/1-marta-edinyj-den-konsultacij-po-voprosam-otnosyashhimsya-k-polnomochiyam-rosreestra/', '')
```

```python
def define_date(date_tuple):
    if date_tuple[1] == 'января':
        date = date_tuple[0] + '.01.' + date_tuple[2]
    if date_tuple[1] == 'февраля':
        date = date_tuple[0] + '.02.' + date_tuple[2]
    if date_tuple[1] == 'марта':
        date = date_tuple[0] + '.03.' + date_tuple[2]
    if date_tuple[1] == 'апреля':
        date = date_tuple[0] + '.04.' + date_tuple[2]
    if date_tuple[1] == 'мая':
        date = date_tuple[0] + '.05.' + date_tuple[2]
    if date_tuple[1] == 'июня':
        date = date_tuple[0] + '.06.' + date_tuple[2]
    if date_tuple[1] == 'июля':
        date = date_tuple[0] + '.07.' + date_tuple[2]
    if date_tuple[1] == 'августа':
        date = date_tuple[0] + '.08.' + date_tuple[2]
    if date_tuple[1] == 'сентября':
        date = date_tuple[0] + '.09.' + date_tuple[2]
    if date_tuple[1] == 'октября':
        date = date_tuple[0] + '.10.' + date_tuple[2]
    if date_tuple[1] == 'ноября':
        date = date_tuple[0] + '.11.' + date_tuple[2]
    if date_tuple[1] == 'декабря':
        date = date_tuple[0] + '.12.' + date_tuple[2]
    return date
```

```python
def clean_news(news):
    news = news.replace('&#8212;', '—')
    news = news.replace('&#171;', '«')
    news = news.replace('&#187;', '»')
    news = re.sub('#gallery.*?{.+?}', '', news)
    news = re.sub('/\*.*?\*/', '', news)
    news = news.replace('<p>', '\n')
    news = news.replace('<br />', '\n')
    news = re.sub('<.+?>', '', news)
    news = re.sub('\s\s(\s)+', '\n', news)
    return news
```

```python
def write_right_news(link, filename):
    r = requests.get(link, cookies=cookies).text
    r = r.replace('\n', '')
    
    date_pattern = 'Написано\: (\d+) ([а-я]+) (\d+) автор\:'
    date_tuple = re.findall(date_pattern, r)[0]
    date = define_date(date_tuple)
    
    article_name = ''
    article_pattern = '<h1>(.+?)</h1>'
    res = re.findall(article_pattern, r)
    if len(res) > 0:
        article_name = res[0]
    
    
    prepattern = '<div id=.+?</div>'
    r = re.sub(prepattern, '', r)
    
    pattern = '<div class="entry">(.+?)</div>'
    news = re.findall(pattern, r)[0]
    news = clean_news(news)
    
    with open(filename, "a", encoding="utf-8") as f:
        string = '=====\n'+link+'\nСельская правда\n' + date + '\nРедакция\n' + article_name + '\n' + news + '\n'
        f.write(string)
        
write_right_news('http://sp-18.ru/2017/03/29-marta-2017-goda-v-regionalnoj-obshhestvennoj-priemnoj-predsedatelya-partii-edinaya-rossiya-d-a-medvedeva-deputat-gosdumy-rf-andrej-isaev-provyol-priyom-grazhdan/', 'qwe1_news.txt')
```

Эта функция перебирает все новости на всех страницах и вызывает предыдущую

```python
def find_all_news(link, filename):
    r = requests.get(link, cookies=cookies).text
    pattern = 'Стр\. 1 из (\d+)'
    page_num = int(re.findall(pattern, r)[0])
    print(page_num)
    for i in range(1, page_num + 1):
        page_link = link + 'page/' + str(i) + '/'
        page_r = requests.get(page_link, cookies=cookies).text
        pattern = '<h2><a title.+?href="(.+?)"'
        news_pages = re.findall(pattern, page_r)
        print(news_pages)
        for news_link in news_pages:
            write_right_news(news_link, filename)
            
#text = find_all_news('http://sp-18.ru/2016/09/', '')
#print(text)
```

```python
def lemmatize_line(line):
    words = [w.lower() for w in word_tokenize(line) if w.isalpha()]
    lemmas = [morph.parse(w)[0].normal_form for w in words if w not in sw]
    # line = list('авбиьбваьбавьбвьаь') # ['кошка', 'кот', 'ывапр']# все русские слова через пробел
    return lemmas
```

```python
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from pymorphy2 import MorphAnalyzer
from collections import defaultdict
morph = MorphAnalyzer()

n = int(input())
graph = defaultdict(lambda: defaultdict(int))
sw = stopwords.words('russian')

with open('sp_news.txt', "r", encoding="utf-8") as f:
    file = f.readlines()
    for line in file:
        lemmas = lemmatize_line(line)
        #print(lemmas)
        if len(lemmas) > 1:
            for i in range(len(lemmas) - n + 1):
                window = lemmas[i:(i+n)]
                #print(window)
                for i in range(len(window)):
                    for j in range(len(window) - i - 1):
                        if window[i] != window[j+i+1]:
                            graph[window[i]][window[j+i+1]] += 1
                            graph[window[j+i+1]][window[i]] += 1
            #print(graph)

#print(graph)
zipf_words = []
zipf = []
for word, neibs in graph.items():
    freqs = (sorted(neibs.values(), reverse = True)[:2])
    if len(freqs) > 1:
        zipf_i = freqs[0]/freqs[1]
        zipf_words.append([word, zipf_i])
        zipf.append(zipf_i)
zipf_mean = np.mean(zipf)
crit = 2*zipf_mean
res_words = []
for word in zipf_words:
    if word[1] > crit:
        res_words.append(word[0])
print(res_words)
# for word in res_words:
    # print(graph[word])
```
